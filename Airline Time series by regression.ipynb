```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Book11.csv')
```

```python
df.head()
```

```python
df.columns = ['month', 'num_passengers']
```

```python
plt.plot(df.num_passengers)
plt.show()
```

```python
series = df.num_passengers.as_matrix()
```

```python
series
```

```python
N = len(series)
ftest = []
for D in range (2,31):
    n = N - D
    X = np.zeros((n, D))
    print(X.shape)
    for d in range(D):
        X[:,d] = series[d:d+n]
        
    Y = series[D:D+n]
    print("X:", X.shape)
    print("Y:", Y.shape)

    print ("series length:", n)
    Xtrain = X[:n//2]
    Ytrain = Y[:n//2]
    Xtest = X[n//2:]
    Ytest = Y[n//2:]

    model = LinearRegression()
    model.fit(Xtrain, Ytrain)
    print ("train score:", model.score(Xtrain, Ytrain))
    print ("test score:", model.score(Xtest, Ytest))
    ftest.append(model.score(Xtest, Ytest))

    # plot the prediction with true values
    plt.plot(series, label = "original")

    train_series = np.empty(n)
    train_series[:n//2] = model.predict(Xtrain) 
    train_series[n//2:] = np.nan
    # prepend d nan's since the train series is only of size N - D
    plt.plot(np.concatenate([np.full(d, np.nan), train_series]), label = "Train")

    test_series = np.empty(n)
    test_series[:n//2] = np.nan
    test_series[n//2:] = model.predict(Xtest)
    plt.plot(np.concatenate([np.full(d, np.nan), test_series]), label = "Test")
    plt.legend()

    plt.show()
    
print("test score", ftest)
kk=[abs(x) for x in ftest]
plt.plot(kk)
```
