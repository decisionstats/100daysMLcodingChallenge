```python
from __future__ import print_function, division
from builtins import range
\

import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 50
D =50
```

```python
X=(np.random.random((N,D))-0.5)*10
```

```python
X.shape
```

```python
true_w = np.array([1,0.5,-0.5]+[0]*(D-3))
```

```python
true_w
```

```python
Y=X.dot(true_w)+np.random.randn(N)*50
```

```python
Y.shape
```

```python
C=[]
```

```python
w = np.random.randn(D)/np.sqrt(D)
```

```python
w
```

```python
lr=0.001
```

```python
l1=20
```

```python
for t in range(500):
    Yhat = X.dot(w)
    delta = Yhat - Y
    w = w- lr*(X.T.dot(delta)+l1*np.sign(w))
    
    mse = delta.dot(delta)/N
    C.append(mse)
    
```

```python
plt.plot(C)
```

```python
w
```

```python
plt.plot(true_w, label = 'true')
plt.plot(w, label = 'predicted')
plt.legend()
```
